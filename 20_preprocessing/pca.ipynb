```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
%config InlineBackend.figure_format = "retina"
```

```python
dta = pd.read_csv("../15_modified_data/train_data.csv")
# dta["nonzero_tip"] = np.where(dta.tip != 0, 1, 0)
```

```python
month_map = {
    "January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12,
}
dta["month_int"] = dta["month"].map(month_map)
day_map = {
    "Monday": 1,
    "Tuesday": 2,
    "Wednesday": 3,
    "Thursday": 4,
    "Friday": 5,
    "Saturday": 6,
    "Sunday": 7,
}
dta["day_ofweek_int"] = dta["day_of_week"].map(day_map)
```

```python
dta.columns
```

```python
X = dta[
    [
        "trip_seconds",
        "trip_miles",
        "fare",
        "additional_charges",
        "trips_pooled",
        "max_temp",
        "min_temp",
        "precip",
        "wind_speed",
        "year",
        "hour",
        "month_int",
        "day_ofweek_int",
        "weekend"
    ]
]
```

```python
dta
```

```python
X
```

```python
scaler = StandardScaler()
dta_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2).fit_transform(dta_scaled)
pcaDF = pd.DataFrame(data = pca, columns=["pc1", "pc2"])
pcaDF = pd.concat([pcaDF, dta.nonzero_tip], axis = 1)
pcaDF.head()
```

```python
np.random.seed(1234)
pca_sampled = pcaDF.sample(2000)
```

```python
f, ax = plt.subplots(1, 2, figsize=(20, 8))
plt.rcParams.update({"font.size": 19})
ax[0].spines['top'].set_visible(False)
ax[0].spines['right'].set_visible(False)
plot = ax[0].scatter(
    pca_sampled.pc1, pca_sampled.pc2, c=pca_sampled.nonzero_tip, cmap="tab20b", s=18
)
ax[0].legend(
    handles=plot.legend_elements()[0], labels=["No Tip", "Tip"], frameon=False, handletextpad=0.05, markerscale=2
)
ax[0].set_title("First Two Principal Components")
ax[0].set_xlabel("Principal Component 1")
ax[0].set_ylabel("Principal Component 2")
pca = PCA().fit(X)
var = np.cumsum(
    pca.explained_variance_ratio_,
)
ax[1].spines['top'].set_visible(False)
ax[1].spines['right'].set_visible(False)
ax[1].plot(list(range(1, 15)), var, marker="o", markersize=4)
ax[1].set_xlabel("Number of Principal Components")
ax[1].set_ylabel("Fraction of Total Variance Explained")
ax[1].set_title("Cumulative Fraction of Variance Explained")
plt.savefig("../26_images/both_pca.png", dpi = 400, bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(8,6))
scatter = plt.scatter(pca_sampled.pc1, pca_sampled.pc2, c = pca_sampled.nonzero_tip, cmap="tab20b", s=14)
plt.legend(handles=scatter.legend_elements()[0], labels=["No Tip", "Tip"], fancybox=False)
plt.title("Tip Distribution after PCA Transformation")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.tight_layout()
plt.savefig("../26_images/pca_2.png", dpi = 200, bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(8,6))
pca = PCA().fit(X)
var = np.cumsum(pca.explained_variance_ratio_)
plt.plot(range(1,15),var, marker = "o", markersize=4)
plt.xlabel("Number of Principal Components")
plt.ylabel("Fraction of Total Variance Explained")
plt.title("Cumulative Fraction of Variance Explained")
plt.savefig("../26_images/pca_all.png", dpi = 200)

plt.show()
```
