```python
import pandas as pd
import numpy as np
```

```python
# read in weather files
weather1 = pd.read_csv("../10_original_data/weather.csv")
weather2 = pd.read_csv("../10_original_data/weather20212022.csv")
```

```python
weather = pd.concat([weather1, weather2], ignore_index=True)
weather
```

```python
# Read in all uber data
df = pd.concat(
    map(
        pd.read_csv,
        [
            "../10_original_data/2020apr_july.csv",
            "../10_original_data/2021may_aug.csv",
        ],
    ),
    ignore_index=True,
)

# Read in weather data

weather["date"] = pd.DatetimeIndex(weather["date"])
weather["merge"] = weather["date"].dt.strftime("%Y/%m/%d")
df["trip_start_timestamp"] = pd.DatetimeIndex(df["trip_start_timestamp"])
df["merge"] = df["trip_start_timestamp"].dt.strftime("%Y/%m/%d")
weather = weather.fillna(0)

# Merge weather data
new = pd.merge(df, weather, how="left", on="merge")
```

```python
pd.set_option('display.max_columns', None)
new
```

```python
new = new.drop(
    [
        "Unnamed: 0_x",
        #"Unnamed: 0.1",
        "Unnamed: 0_y",
        "merge",
        "date_y",
    ],
    axis=1,
)
new = new.rename(columns={"date_x": "date"})

new.to_csv("../15_modified_data/post_covid_final.csv")
```
